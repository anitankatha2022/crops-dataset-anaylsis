# Data vizualisation

```python
plantpart_counts = { key:i for key, i in zip(clo_data.PlantPartHarvested.value_counts().index[:5], range(5)) }

def get_plantpart_classif(val):
    try:
        plantpart_counts[val]
        return val
    except:
        return "Others"
    

clo_data['PlantPartClassif'] = clo_data['PlantPartHarvested'].apply(get_plantpart_classif)
clo_data.PlantPartClassif.value_counts()
```

```python
plt.style.use("seaborn-white")

theme = plt.get_cmap('bwr')

fig, ax = plt.subplots(figsize=(10, 10))
wedgeprops = {'width':0.3, 'edgecolor':'white', 'linewidth':3}
ax.set_prop_cycle("color", [theme(1. * i / len(plantpart_counts))
                             for i in range(len(plantpart_counts))])
ax.pie(clo_data.PlantPartClassif.value_counts(), wedgeprops=wedgeprops, startangle=90, autopct='%.0f%%')
plt.text(0, 0, "Plant Part Harvested", ha='center', va='center', fontsize=20)
plt.legend(clo_data.PlantPartClassif.unique(), prop={'size': 11})
plt.show()
```

```python
cropcategory_counts = clo_data['CropCategory'].value_counts()

fig, ax = plt.subplots(figsize=(10, 10))
wedgeprops = {'width':0.3, 'edgecolor':'white', 'linewidth':3}
ax.set_prop_cycle("color", [theme(1. * i / len(cropcategory_counts))
                             for i in range(len(cropcategory_counts))])
ax.pie(clo_data.CropCategory.value_counts(), wedgeprops=wedgeprops, startangle=90, autopct='%.0f%%')
plt.text(0, 0, "Crop Category", ha='center', va='center', fontsize=20)
plt.legend(cropcategory_counts.index , prop={'size': 10})
plt.show()
```

```python
def plot_strip_box(x, y):
    """
    Create a strip plot and box plot for the given x and y values
    """
    
    fig= plt.subplots(figsize=(20, 10))
    plt.subplots_adjust(hspace=0.3)

    ax1 = plt.subplot(1, 2, 1)
    sns.stripplot(x = clo_data[x], y = clo_data[y])
    plt.tick_params(rotation=60, labelsize=10)

    ax2 = plt.subplot(1, 2, 2)
    sns.boxplot(x = clo_data[x], y = clo_data[y])
    plt.tick_params(rotation=60, labelsize=10)

    plt.suptitle("{} of different {}".format(x,y), fontsize=20)
    plt.show()
```

# Train and Test

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# selecting the needed columns for this part
final_df = clo_data.drop('PlantPartClassif',axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
X = final_df[[
 'AvMoisture%',
 'AvN%(dry)',
 'AvP%(dry)',
 'AvK%(dry)',
 ]]
y = final_df['CropCategory']


X_cols = X.columns
#scaling feature values
scaler = StandardScaler()
X = scaler.fit_transform(X)

# returning scaled values to dataframe
X = pd.DataFrame(X, columns=X_cols)
```

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()
y = pd.Series(enc.fit_transform(y))
y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)
```

# xgboost

```python
import xgboost as xgb
from sklearn.utils.class_weight import compute_sample_weight

sample_weights = compute_sample_weight(
    class_weight='balanced',
    y=y_train
)

xg_reg_cc = xgb.XGBClassifier(learning_rate=0.1,
                               n_estimators=1000,
                    max_depth=5,
                    min_child_weight=1,
                    gamma=0,
                    subsample=0.8,
                    colsample_bytree=0.8,
                    objective='multi:softmax',
                    eval_metric='merror',
                    nthread=4,
                               num_class=9,
                    seed=27)
xg_reg_cc.fit(X_train, y_train, sample_weight=sample_weights)
preds = np.array(xg_reg_cc.predict(X_test))
print(classification_report(preds, y_test))
```

```python
f, ax = plt.subplots(figsize=(10,5))
plot = sns.barplot(x=X.columns, y=xg_reg_cc.feature_importances_)
ax.set_title('XGBoost Feature Importance')
plot.set_xticklabels(plot.get_xticklabels(),rotation='vertical')
plt.show()
```

# GradientBoostingClassifier

```python
from sklearn.ensemble import GradientBoostingClassifier
grad_cc = GradientBoostingClassifier().fit(X_train, y_train)
print(classification_report(grad_cc.predict(X_test),y_test))
```

# DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier

dtree_cc = DecisionTreeClassifier(random_state=42).fit(X_train, y_train)
print(classification_report(dtree_cc.predict(X_test),y_test))
```

```python
plt.figure(figsize=(10,5), dpi=80)
tree_features = len(X_train.columns)
plt.barh(range(tree_features), dtree_cc.feature_importances_)
plt.xlabel("Feature importance")
plt.ylabel("Feature name")
plt.yticks(np.arange(tree_features), X_train.columns)
plt.show()
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
forest_cc = RandomForestClassifier(max_depth=5,n_estimators=100,random_state=42).fit(X_train, y_train)

print(classification_report(forest_cc.predict(X_test), y_test))
```
